```python
import numpy as np
from numpy.linalg import norm, inv
import numpy.linalg as la
import re
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

verySmallNumber = 1e-14
def gsBasis(A) :
    B = np.array(A, dtype=np.float_) # Make B as a copy of A, since we're going to alter it's values.
    for i in range(B.shape[1]) :
        for j in range(i) :
            B[:, i] =  B[:, i] - B[:, i] @ B[:, j] * B[:, j]
        if la.norm(B[:, i]) > verySmallNumber :
            B[:, i] = B[:, i] / la.norm(B[:, i])
        else :
            B[:, i] = np.zeros_like(B[:, i])   
      
    return B
a,b=[],[]
n=int(input('enter the num of row ... '))

for i in range(n):
    x=list(map(int,input().split()))
    a.append(x)
a=np.array(a)
print('The matrix generated successfully \ncall this matrix a')

new_matrix=input('do you will need anouther matrix? y/n ... ')
new_matrix=new_matrix.lower()
if (new_matrix=='y'):
    n=int(input('enter the num of row ... '))

    for p in range(n):
        y=list(map(int,input().split()))
        b.append(y)
    b=np.array(b)
    print('The matrix generated successfully \n call this matrix b')

print('\nwelcome to our program we provide a lot of process like:-\n\
1.add and sub Matrices. to do that write like that x+b ... x,y are Matrices\n\
2. dot product write it like this x@y ... a,y are Matrices\n\
3.scale any matrix write (matrix you want)*scale(number)\n\
4.to get the Basis for matrix write Basis (matrix you want)\n\
5. to transpose matrix wrie transpose (matrix you want)\n\
6. to get inverse for matrix write inverse (matrix you want)')
process=input()
process= process.split()
for i in process:
    if (i=='a'):
        process[process.index(i)]=a
    if (i=='b'):
        process[process.index(i)]=b
for i in process:        
    if (i=='+'):
        print(process[process.index(i)-1]+process[process.index(i)+1])
    if (i=='-'):
         print(process[process.index(i)-1]-process[process.index(i)+1])
    if (i=='*'):
        print(process[process.index(i)-1]*int(process[process.index(i)+1]))
    if (i=='@'):
        print(process[process.index(i)-1]@process[process.index(i)+1])
    if (i=='Basis'):
        print(gsBasis(process[process.index(i)+1]))
    if (i=='inverse'):
         print(inv(process[process.index(i)+1]))
    if (i=='transpose'):
         print(np.transpose(process[process.index(i)+1]))
anouther=input('do you want anouther ? y/n... ')
while(anouther=='y'):
    process=input()
    process= process.split()
    for i in process:
        if (i=='a'):
            process[process.index(i)]=a
        if (i=='b'):
            process[process.index(i)]=b
    for i in process:        
        if (i=='+'):
            print(process[process.index(i)-1]+process[process.index(i)+1])
        if (i=='-'):
            print(process[process.index(i)-1]-process[process.index(i)+1])
        if (i=='*'):
            print(process[process.index(i)-1]*int(process[process.index(i)+1]))
        if (i=='@'):
            print(process[process.index(i)-1]@process[process.index(i)+1])
        if (i=='Basis'):
            print(gsBasis(process[process.index(i)+1]))
        if (i=='inverse'):
            print(inv(process[process.index(i)+1]))
        if (i=='transpose'):
            print(np.transpose(process[process.index(i)+1]))
    anouther=input('do you want anouther ? y/n.. ')
print('Thanks for using our program, we are happy to serve you')
```
